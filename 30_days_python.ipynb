## Day 28 — Classes & Objects (Rafiki’s Character)


```python
class Player:
    def __init__(self, name, game, level):
        self.name = name
        self.game = game
        self.level = level

    def status(self):
        return f"{self.name} is level {self.level} in {self.game}"

rafiki = Player("Rafiki", "FC26", 15)
print(rafiki.status())
```

## Day 29 — Inheritance (Special Player Type)


```python
class ProPlayer(Player):
    def __init__(self, name, game, level, rank):
        super().__init__(name, game, level)
        self.rank = rank

    def status(self):
        return f"{self.name} is rank {self.rank} and level {self.level} in {self.game}"

pro = ProPlayer("Rafiki", "FC26", 15, "Gold")
print(pro.status())
```

## Day 30 — Final Project
*"Tie it all together — simulate your life in Python."*

```python
import random

def energy_boost(hours_sleep, coffee):
    return hours_sleep * 10 + coffee * 5

class RafikiLife:
    def __init__(self):
        self.mood = random.choice(["Chill", "Stressed", "Motivated"])
        self.energy = energy_boost(5, 1)
        self.gaming = True

    def status(self):
        game_status = "Playing FC26" if self.gaming else "No gaming today"
        return f"Mood: {self.mood}, Energy: {self.energy}, {game_status}"

my_day = RafikiLife()
print(my_day.status())
```
